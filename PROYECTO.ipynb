# Usaremos knn para columnas numericas

```python
#Columna Budget KNN
imputer= KNNImputer(n_neighbors=5)
# Creamos un array para especificar las columnas que queremos imputar
columnas=['Budget', 'Revenue', 'Spend', 'Impressions','CTR', 'Conversions','CR', 'ROI', 'Duration', 'Clicks', 'Customer_Satisfaction', 'Campaign_Effectiveness']
# Aplicamos el imputer
df[columnas]= imputer.fit_transform(df[columnas])
print(df.isnull().sum())
```

```python
#DECLARAMOS LOS CUARTILES
Q1= df['Revenue'].quantile(0.25)
Q3= df['Revenue'].quantile(0.75)
IQR= Q3-Q1
#cALCULAMOS LOS LIMITES INFERIOR Y SUPERIOR PARA VER SI SE PASAN Y SON OUTLIERS
LIMITEINFERIOR= Q1-1.5*IQR
LIMITESUPERIOR= Q3+1.5*IQR
print(LIMITEINFERIOR, LIMITESUPERIOR)
outliers= df[(df['Revenue']<LIMITEINFERIOR) | (df['Revenue']>LIMITESUPERIOR)]
print(len(outliers))

Q1= df['Budget'].quantile(0.25)
Q3= df['Budget'].quantile(0.75)
IQR= Q3-Q1
LIMITEINFERIOR= Q1-1.5*IQR
LIMITESUPERIOR= Q3+1.5*IQR
print(LIMITEINFERIOR, LIMITESUPERIOR)
outliers= df[(df['Budget']<LIMITEINFERIOR) | (df['Budget']>LIMITESUPERIOR)]
print(len(outliers))

Q1= df['Spend'].quantile(0.25)
Q3= df['Spend'].quantile(0.75)
IQR= Q3-Q1
LIMITEINFERIOR= Q1-1.5*IQR
LIMITESUPERIOR= Q3+1.5*IQR
print(LIMITEINFERIOR, LIMITESUPERIOR)
outliers= df[(df['Spend']<LIMITEINFERIOR) | (df['Spend']>LIMITESUPERIOR)]
print(len(outliers))


```

# Analisis de clubster

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler= StandardScaler()
df_num= df.select_dtypes(include=['float64', 'int64'])
df_num.head()
df_scale= scaler.fit_transform(df_num)

#Metodo de codo para la dettarminacion de clusters
from sklearn.cluster import KMeans
codo=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scale)
    codo.append(kmeans.inertia_)

plt.plot(range(1,11), codo, marker='o')
plt.show()

```
